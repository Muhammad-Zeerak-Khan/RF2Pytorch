```python
from sklearn import datasets
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import hummingbird.ml 
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
# Breast Cancer Dataset
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
#Create pandas dataframe

df = pd.DataFrame(data["data"],columns = data["feature_names"])
df['target'] = data['target']
df.head()
```

```python
X3 = df.drop(columns='target')
y3 = df['target']
```

```python
x_train , x_test, y_train,y_test = train_test_split(X3, y3, test_size = .25, random_state = 30)
```

```python
rf = RandomForestClassifier(n_estimators=10)
```

```python
rf.fit(x_train, y_train)
```

```python
rf.score(x_test, y_test)
```

```python
rf.estimators_
```

```python
grid_pram = {
    "n_estimators" : [5,10 , 50 , 100 , 120 , 150],
    'criterion' :['gini' ,'entropy'],
    'max_depth' :range(10),
    'min_samples_leaf' :range(10)
    
}
```

```python
grid_serach_rf = GridSearchCV(param_grid= grid_pram, cv = 5 , n_jobs=6,verbose=1 ,estimator = rf)
```

```python
grid_serach_rf.fit(x_train, y_train)
```

```python
grid_serach_rf.best_params_
```

```python
grid_serach_rf.best_score_
```
